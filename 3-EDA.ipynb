```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import json
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

import plotly.graph_objects  as go

```

```python
df=pd.read_csv("./Bank_nifty_data.csv")
```

```python
df.head()
```

```python
def change(data):
    value =""
    for i in data:
        if i == ",":
            continue
        else:
            value +=i
    return float(value)
```

```python
def changevalume(data):
    value =""
    if "M" in data:
        for i in data:
            if i == 'M':
                continue
            else:
                value += i
        return int(float(value)*1000000)
    elif "B" in data:
        for i in data:
            if i == 'B':
                continue
            else:
                value += i
        return int(float(value)*1000000000)
    elif data== 0:
        return 0
    else:
        return "M"
```

```python
def changechag(data):
    if "%" in data:
      return data[:-1]
    else:
        return "M"
```

```python
df["Close"] =df["Close"].apply(change)
df["Open"] =df["Open"].apply(change)
df["High"] =df["High"].apply(change)
df["Low"] =df["Low"].apply(change)
```

```python
df["Volume"] = df["Volume"].apply(changevalume)
```

```python
df["Chg%"] =df["Chg%"].apply(changechag)
```

```python
df.drop(df[df['Volume'] == "M"].index, inplace = True)
```

```python
df.head()
```

```python
df["Date"] =pd.to_datetime(df["Date"])
```

```python
df.set_index(pd.DatetimeIndex(df['Date']),inplace=True)
```
